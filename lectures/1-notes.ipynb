## Lecture - 2nd Hour

The history of AI:

* Generation #1: Good Old-Fashioned AI
    * Handcraft predictions
    * Learn nothing
* Generation #2: Shallow Learning
    * Handcraft features
    * Learn predictions
* Generation #3: Deep Learning
    * Handcraft algorithm
    * Learn features and predictions end-to-end
* Generation #4: Meta Learning
    * Handcraft nothing
    * Learn algorithm and features and predictions end-to-end

We focus on generation #2 and #3.

### ML problem types

* Supervised learning
    * Provided with $\{x_i, y_i\}_{i = 1}^{N}$
    * Train using input and correct solution, use model to predict unknown solutions
* Usnupervised learning
    * Provided with $\{x_i\}_{i=1}^{N}
    * Clustering for instance
* Reinforcement learning
    * Provided with feedback, a computer program interacts with a dynamic environment to achieve a certain goal
    * Super mario for instance
    
"CRoss Industry Standard Process for Data Mining", the CRISP-DM model, is used to guide the development of machine learning algorithms.

Multidimensional data

* Data warehouse uses multidimensional data
* d = 1 -> vector, d = 2 -> matrix, d = 3 -> tensor
* Most common format for unstructured data
* Pandas support d <= 3
* Tabular data is widely used in ML (and we will use it in this course)

Pandas data structure

* Series is vector or 1D array
* DataFrame is a matrix, table, or 2D array, container of Series
* Panel is a cube or }D array, container of DataFrame

Data interface - extraction, transformation, and loading (ETL)

* Data extraction

* Load - Sort, summarize, consolidate, compute views, check integrity, and build indicies and partitions
